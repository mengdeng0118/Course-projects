```python
import sys
import string 
import pandas as pd
from operator import itemgetter
from collections import defaultdict
from collections import Counter
import numpy as np
import math
import random
import matplotlib.pyplot as plt
%matplotlib inline

plt.interactive(True)
```

# 0. Preprocessing the file

```python
def process_str(s):
    return s.translate(str.maketrans('', '', string.punctuation)).lower().split()

# dataset format:
# list of (class_label, set of words)
def read_dataset(file_name):
    dataset = []
    with open(file_name) as f:
        for line in f:
            st = line.strip().split('\t')
            if len(st) == 3:
                index = st[0]
                class_label = st[1]
                text = st[2]
                words = process_str(text)
                dataset.append( (int(class_label), set(words)) )
    return dataset
```

# 1. Build the feature dictionary based on the given training data. 

```python
def get_most_commons(dataset, skip=100, total=100):
    my_list = []
    for item in dataset:
        my_list += list(item[1])

    counter = Counter(my_list)

    temp = counter.most_common(total+skip)[skip:]

    words = [item[0] for item in temp]
    return words

# the length of the common words will be the
# length of the vectors
def generate_vectors(dataset, common_words):
    d = {}
    for i in range(len(common_words)):
        d[common_words[i]] = i
    
    vectors = []
    for item in dataset:
        vector = [0] * len(common_words)
        for word in item[1]:
            if word in d:
                vector[d[word]] = 1

        vectors.append( (item[0], vector) )

    return vectors
```

# 2. Implement Decision Trees

```python
#Given a splited left and right nodes, and the splitted value, return the Gini evaluation.
def gini_index(x_matrix, y):
    col = x_matrix.shape[1]
    row = x_matrix.shape[0]
    
    gini1 = []
    gini0 = []
    
    for i in range(col):
        
        y1 = y[np.where(x_matrix[:, i] == 1)]
        y0 = y[np.where(x_matrix[:, i] == 0)]
        
        if len(y1) == 0:
            gini1.append(0)
        else:
            p1 = sum(y1)/len(y1)
            gini1.append((1-p1**2-(1-p1)**2)*len(y1)/len(y))
        
        if len(y0) == 0:
            gini0.append(0)
        else:
            p0 = sum(y0)/len(y0)
            gini0.append((1-p0**2-(1-p0)**2)*len(y0)/len(y))
    
    return np.asarray(gini1), np.asarray(gini0)


def gini_gain(data_matrix):
    #First calculate gini_index based on the input y_obs
    y_obs = data_matrix[:, -1]

    freq_y_obs = (np.array([sum(y_obs), y_obs.shape[0] - sum(y_obs)]) / y_obs.shape[0])
    
    gini_y = 1 - sum(np.power(freq_y_obs, 2))
    
    x_matrix = data_matrix[:, 0:(data_matrix.shape[1]-1)]
    
    gini1, gini0 = gini_index(x_matrix, y_obs)
    
    giniGain = gini_y - gini1 - gini0
        
    return giniGain


#Given a dataset, split a node, return two splitted groups as two arrays.
def split(idx, data_matrix):
    
    left = data_matrix[data_matrix[:,idx]==0,:]
    right = data_matrix[data_matrix[:,idx]==1,:]
    
    return left, right
   

#Given a matrix, evaluate all possible splits, and return a dict with the index of the split, best GiniGain, and left/right groups.
def get_split(data_matrix):
    
    #If y_obs all belong to one class, then stop splitting. 
    
    #get the gini index of the max giniGain across all feature columns. When ties happen, just pick the first one.
    best_split_idx = np.argmax(gini_gain(data_matrix))
    
    best_gain = np.max(gini_gain(data_matrix))
    
    #Split only once, what is the score here? 
    groups = split(best_split_idx, data_matrix)
    if groups[0].shape[0] == 0:
        return {'index': best_split_idx, 'score': best_gain, 'groups': (None, groups[1])}
    if groups[1].shape[0] == 0:
        return {'index': best_split_idx, 'score': best_gain, 'groups': (groups[0], None)}
    
    #store the split object in a dictionary, note groups include the y column.
    return {'index': best_split_idx, 'score': best_gain, 'groups': groups}

#Build a tree. 

#Stops when the current group has only one class.
def to_terminal(data_matrix):
    outcomes = np.bincount(data_matrix[:,-1])
    return np.argmax(outcomes)

def recursive_split(node, max_depth, min_size, depth):
    
    if not isinstance(node, dict):
        return
    
    left, right = node['groups']
    del(node['groups'])
    
    if depth >= max_depth:
        node['left'], node['right'] = to_terminal(left), to_terminal(right)
        return
   
    if left is None or left.shape[0] < min_size or sum(left[:,-1]) == left.shape[0] or sum(left[:,-1]) == 0:
        node['left'] = to_terminal(left)
    else:
        node['left'] = get_split(left) 
        recursive_split(node['left'], max_depth, min_size, depth+1)
        
    if right is None or right.shape[0] < min_size or sum(right[:,-1]) == right.shape[0] or sum(right[:,-1]) == 0:
        node['right'] = to_terminal(right)
    else:
        node['right'] = get_split(right)
        recursive_split(node['right'], max_depth, min_size, depth+1)
          
        
def decision_tree_learn(data_matrix, max_depth, min_size):
    root = get_split(data_matrix)
    #print_tree(root,0)
    recursive_split(root, max_depth, min_size, 1)
    return root

# Print a decision tree
def print_tree(node, depth=0):
    if isinstance(node, dict):
        print('%s[X%d]' % ((depth*' ', (node['index']+1))))
        print_tree(node['left'], depth+1)
        print_tree(node['right'], depth+1)
    else:
        print('%s[%s]' % ((depth*' ', node)))

# Make a prediction
def predict(node, row):
        if row[node['index']] == 0:
            if isinstance(node['left'], dict):
                return predict(node['left'], row)
            else:
                return node['left']
        else:
            if isinstance(node['right'], dict):
                return predict(node['right'], row)
            else:
                return node['right']


def decision_tree_predict(train, test, max_depth, min_size):
    tree = decision_tree_learn(train, max_depth, min_size)
    predictions = list()
    for i in range(test.shape[0]):
        predictions.append(predict(tree, test[i,:]))
    return(predictions)

def zero_one_loss(y_pred, test):
    
    y_obs = test[:, -1].tolist()
    
    compare_result = np.where(np.not_equal(y_obs, y_pred), 1, 0)
    
    return sum(compare_result) / len(y_obs)
```

# 3. implement bagging.

```python
random.seed(123)

def sample(train):
    sampleInd = np.random.choice(range(train.shape[0]), train.shape[0], replace=True)
    subsample = train[sampleInd,:]
    return subsample


def bagging_pred(train, test, max_depth=10, min_size=10, tree_number=50):
    
    bags = np.zeros(test.shape[0])
    
    for i in range(tree_number):
        subsample = sample(train)
        predict = decision_tree_predict(subsample, test, max_depth, min_size)                   
        bags += predict
    bags = 0 + (bags > (tree_number / 2))     
    return(bags)
    
    
```

# 4. Implement Random Forest.

```python
# Random Forest Algorithm
def get_split_RT(data_matrix):
    
    n_features = int(math.sqrt(data_matrix.shape[1] - 1))
    attributes_selected = np.random.choice(range(data_matrix.shape[1] - 1), n_features, replace=False)
    attributes_selected = np.append(attributes_selected, -1)
    #form new matrix with selected asttributes.
    new_train = data_matrix[:, attributes_selected]
    #get the gini index of the max giniGain across all feature columns. When ties happen, just pick the first one.
    best_split_idx = attributes_selected[np.argmax(gini_gain(new_train))]
   
    best_gain = np.max(gini_gain(new_train))
    
    #Split only once, what is the score here? 
    groups = split(best_split_idx, data_matrix)
    
    if groups[0].shape[0] == 0:
        return {'index': best_split_idx, 'score': best_gain, 'groups': (None, groups[1])}
    if groups[1].shape[0] == 0:
        return {'index': best_split_idx, 'score': best_gain, 'groups': (groups[0], None)}
    
    #store the split object in a dictionary, note groups include the y column.
    return {'index': best_split_idx, 'score': best_gain, 'groups': groups}        

def recursive_split_RT(node, max_depth, min_size, depth):
    
    if not isinstance(node, dict):
        return
    
    left, right = node['groups']
    del(node['groups'])
    
    if left is None or left.shape[0] == 0:
        if right is None or right.shape[0] == 0:
            return
        else:
            node['left'] = node['right'] = to_terminal(right)
            return
    if right is None or right.shape[0] == 0:
        if left is None or left.shape[0] == 0:
            return
        else:
            node['left'] = node['right'] = to_terminal(left)
            return
    
    if depth >= max_depth:
        if left is None or left.shape[0] == 0 or right is None or right.shape[0] == 0:
            return
        else:
            node['left'], node['right'] = to_terminal(left), to_terminal(right)
            return
   
    if left.shape[0] < min_size or sum(left[:,-1]) == left.shape[0] or sum(left[:,-1]) == 0:
        node['left'] = to_terminal(left)
    else:
        node['left'] = get_split_RT(left)
        if node['left'] is None:
            node['left'] = to_terminal(left)
        else:
            recursive_split_RT(node['left'], max_depth, min_size, depth+1)
        
    if right.shape[0] < min_size or sum(right[:,-1]) == right.shape[0] or sum(right[:,-1]) == 0:
        node['right'] = to_terminal(right)
    else:
        node['right'] = get_split_RT(right)
        if node['right'] is None:
            node['right'] = to_terminal(right)
        else:
            recursive_split_RT(node['right'], max_depth, min_size, depth+1)
        
def random_forest_learn(train, max_depth, min_size):
    root = get_split_RT(train)
    recursive_split_RT(root, max_depth, min_size, 1)
    return root
    

def random_forest_predict(train, test, max_depth, min_size):
    tree = random_forest_learn(train, max_depth, min_size)
    predictions = list()
    for i in range(test.shape[0]):
        predictions.append(predict(tree, test[i,:]))
    return(predictions)
    
def random_forest(train, test, max_depth=10, min_size=10, tree_number=50):
    bags = np.zeros(test.shape[0])
    for i in range(tree_number):
        subsample = sample(train)
        predict = random_forest_predict(subsample, test, max_depth, min_size)
        bags += predict
    bags = 0 + (bags > (tree_number / 2)) 
    return(bags)
```

# 5. Implement SVM.

```python
##Give weights parameters, return the gradients based on that weights.
def svm_learn(weights, train_feature_matrix, feature_number, lamda, eta_svm, epsilon, iterator):
    ###initialize w and the gradient of w.
    length = len(train_feature_matrix)
   
    #a Read in training data and learn the logistic regression model.
    y_obs = [l[0] for l in train_feature_matrix]


    ##append it as a column of feature_matrix.
    df_class_neg = np.copy(y_obs) 
    df_class_neg[np.where(df_class_neg==0)] = -1

    #get only the xs from the freature matrix
    x_feature = [l[1] for l in train_feature_matrix]
    
    ###Append x_0 = 1 to the x vector.
    x_0 = np.array([1]*(length))
    x = np.append(x_0[:, None], x_feature, 1)
    
    ###calculate gradient for each parameter.
    diff = 100
    i = 0
    
    new_w = weights
    
    while (diff > epsilon and i < iterator):
        #print(i)
        old_w = new_w
        #print(old_w)
        
        #Use dot product
        y_est = np.dot(x, old_w)  
        
        x_trans = np.matrix.transpose(x) #matrix with feature_number*N
        
        y_y_est = df_class_neg * y_est #y*y_est, length

        nabla_hask = np.where(y_y_est < 1, 1, 0) #vector with length*1
        
        y_nabla_hask = df_class_neg * nabla_hask #vector with length*1
        
        y_hask_x_trans = np.dot(x_trans, y_nabla_hask) #vector with feature_number*1
        
        lamda_w = np.dot(lamda, old_w) #feature_number*1

        y_hask_x_trans_N = np.dot(1.0/length, y_hask_x_trans)
        
        gradient = lamda_w - y_hask_x_trans_N
        new_w = old_w - np.dot(eta_svm, gradient)

        i = 1 + i
        
    return new_w

def svm_classify(new_weights, test_feature_matrix, feature_number):

    ###Append x_0 = 1 to the x vector in the test dataset.
    length = len(test_feature_matrix)
    
    #get only the xs from the freature matrix
    x_feature = [l[1] for l in test_feature_matrix]
    
    x_0 = np.array([1]*(length))
    x = np.append(x_0[:, None], x_feature, 1)
    
    y_hat = np.dot(x, new_weights)
    
    y_pred = np.where((y_hat > 0), 1, 0)

    return y_pred

def zero_one_loss2(y_pred, feature_matrix):
    
    length = len(feature_matrix)
    
    #a Read in training data and learn the logistic regression model.
    y_obs = [l[0] for l in feature_matrix]
    
    compare_result = np.where(np.not_equal(y_obs, y_pred), 1, 0)

    
    return sum(compare_result) / length
```

# 6. Boosting.

```python
# Boosting Algorithm
def gini_index_boosting(x_matrix, y, w):
   
    col = x_matrix.shape[1]
    row = x_matrix.shape[0]
        
    gini1 = []
    gini0 = []
    
    for i in range(col):

        y1 = y[np.where(x_matrix[:, i] == 1)]
        y0 = y[np.where(x_matrix[:, i] == 0)]
        
        w1 = w[np.where(x_matrix[:,i] == 1)]
        w0 = w[np.where(x_matrix[:,i] == 0)]

        if len(y1) == 0:
            gini1.append(0)
        else:
            p1 = np.inner(y1, w1)/sum(w1)
            gini1.append(1-np.power(p1,2)-np.power(1-p1,2))
        
        if len(y0) == 0:
            gini0.append(0)
        else:
            p0 = np.inner(y0, w0)/sum(w0)
            gini0.append(1-np.power(p0,2)-np.power(1-p0,2))
    
    return np.asarray(gini1), np.asarray(gini0), w1, w0

def gini_gain_boosting(data_matrix, w):
    #First calculate gini_index based on the input y_obs
    y_obs = data_matrix[:, -1]
    
    p1 = np.inner(y_obs, w)/sum(w)
    
    gini_y = 1 - np.power(p1, 2) - np.power(1-p1, 2)
    
    x_matrix = data_matrix[:, 0:(data_matrix.shape[1]-1)]
    
    gini1, gini0, w1, w0 = gini_index_boosting(x_matrix, y_obs, w)
    
    giniGain = gini_y - gini1*sum(w1)/sum(w) - gini0*sum(w0)/sum(w)

    return giniGain

#Given a matrix, evaluate all possible splits, return a dict with 
    # 1) the index of the split, 
    # 2) best GiniGain, and
    # 3) left/right groups.
def get_split_boosting(data_matrix, w):
    
    #If y_obs all belong to one class, then stop splitting. 
    
    #get the gini index of the max giniGain across all feature columns. When ties happen, just pick the first one.
    best_split_idx = np.argmax(gini_gain_boosting(data_matrix, w))
    
    best_gain = np.max(gini_gain_boosting(data_matrix, w))
    
    #Split only once, what is the score here? 
    groups = split_boosting(best_split_idx, data_matrix, w)
    
    if groups[0].shape[0] == 0:
        return {'index': best_split_idx, 'score': best_gain, 'groups': (None, None, groups[2], groups[3])}
    if groups[1].shape[0] == 0:
        return {'index': best_split_idx, 'score': best_gain, 'groups': (groups[0], groups[1], None, None)}
    
    #store the split object in a dictionary, note groups include the y column.
    return {'index': best_split_idx, 'score': best_gain, 'groups': groups}

def split_boosting(idx, data_matrix, weight):
    
    left = data_matrix[data_matrix[:,idx]==0,:]
    right = data_matrix[data_matrix[:,idx]==1,:]
    
    leftw = weight[data_matrix[:,idx]==0]
    rightw = weight[data_matrix[:,idx]==1]
    
    return left, leftw, right, rightw


#Stops when the current group has only one class.
def to_terminal_boosting(data_matrix, w):
    y = data_matrix[:,-1]
    if np.inner(y, w) >= (sum(w) - np.inner(y, w)):
        return 1
    else:
        return 0

def boosting_learn(data_matrix, weight, max_depth, min_size):
    root = get_split_boosting(data_matrix, weight)
    recursive_split_boosting(root, weight, max_depth, min_size, 1)
    return root


def recursive_split_boosting(node, max_depth, min_size, depth):   
    #base case
    if not isinstance(node, dict):
        return
    
    left, leftw, right, rightw = node['groups']
    del(node['groups'])
    
    
    if left is None or left.shape[0] == 0:
        if right is None or right.shape[0] == 0:
            return
        else:
            node['left'] = node['right'] = to_terminal_boosting(right, rightw)
            return
    if right is None or right.shape[0] == 0:
        if left is None or left.shape[0] == 0:
            return
        else:
            node['left'] = node['right'] = to_terminal_boosting(left, leftw)
            return
    
    if depth >= max_depth:
        if left is None or left.shape[0] == 0 or right is None or right.shape[0] == 0:
            return
        else:
            node['left'] = to_terminal_boosting(left, leftw)
            node['right'] = to_terminal_boosting(right, rightw)
            return
    
    #Non-trivial case
    if left.shape[0] < min_size or sum(left[:,-1]) == left.shape[0] or sum(left[:,-1]) == 0:
        node['left'] = to_terminal_boosting(left, leftw)
    else:
        node['left'] = get_split_boosting(left, leftw)
        if node['left'] is None:
            node['left'] = to_terminal_boosting(left, leftw)
        else:
            recursive_split_boosting(node['left'], max_depth, min_size, depth+1)
        
    if right.shape[0] < min_size or sum(right[:,-1]) == right.shape[0] or sum(right[:,-1]) == 0:
        node['right'] = to_terminal_boosting(right, rightw)
    else:
        node['right'] = get_split_boosting(right, rightw)
        if node['right'] is None:
            node['right'] = to_terminal_boosting(right, rightw)
        else:
            recursive_split_boosting(node['right'], max_depth, min_size, depth+1)

#Build the tree classifer        
def boosting_learn(train, weight, max_depth, min_size):
    
    root = get_split_boosting(train, weight)
    
    recursive_split_boosting(root, max_depth, min_size, 1)
    
    return root

#For each iteration, return prediction, error, and incorrect prediction based on train.
def boosting_predict(train, test, weight, max_depth, min_size):
    
    tree = boosting_learn(train, weight, max_depth, min_size)

    predictions = list()
    
    for i in range(test.shape[0]):
        #Use the tree to predict the test data.
        if (int(predict(tree, test[i,:]) == 1)):
            predictions.append(1)
        elif (int(predict(tree, test[i,:]) == 0)):
            predictions.append(-1)
        else:
            predictions.append(-100)
        
    #Now compute the incorrect predictions based on the train data.
    predictions_train = list()
    for i in range(train.shape[0]):
        predictions_train.append(predict(tree, train[i,:]))
    
    #Use the predictions_train info to calculate the incorrect.
    incorrect = np.where(np.not_equal(train[:,-1], predictions_train), 1, -1) #mark the errors.
    incorrect_idx = np.where(np.not_equal(train[:,-1], predictions_train), 1, 0) #is it necessary?
   
    error = np.dot(weight, incorrect_idx)   

    return(predictions, incorrect, incorrect_idx, error)

    
def boosting(train, test, max_depth=10, min_size=10, tree_number=50):
    
    weight = np.divide(np.ones(train.shape[0]), train.shape[0]) 
    
    pred_test = np.zeros(test.shape[0]) 
    # initial weight
    
    for i in range(tree_number):

        predictions, incorrect, incorrect_idx, error = boosting_predict(train, test, weight, max_depth, min_size) 
        
        alpha = 0.5 * np.log(np.divide(1-error+1e-9, error+1e-9))

        weight = weight*(np.exp(np.multiply(alpha,incorrect_idx)))

        #normalize weights
        weight = weight/sum(weight)
        
        pred_test = pred_test + np.multiply(alpha, predictions)
    
    #Output the sum classifiers.
    pred = np.zeros(len(pred_test))
    
    pred[np.where(pred_test>0)] = 1 
    
    return pred
```

```python
#Plot 1
def main():
    
    filename = 'yelp_data.csv'
    df = read_dataset(filename) # df2 with duplicate, df without duplicate
    
    feature_number = 500
    
    lamda = 0.01
    eta_lr = 0.01
    eta_svm = 0.5
    epsilon = 1e-6
    iterator = 100
    weights = [0]*(feature_number  + 1)

    m = [0.025, 0.05, 0.125, 0.25]
    n = 10
    df_sub = np.array_split(df, n)
    D = len(df) # 2000
    train_sample_size = [int(mm*D) for mm in m] #np.dot(m, D)
    DT_test_zero_one_loss = []
    BT_test_zero_one_loss = []
    RF_test_zero_one_loss = []
    svm_test_zero_one_loss = []
    boosting_test_zero_one_loss = []

    for i in range(len(train_sample_size)): # l = 4
        DT_temp = []
        BT_temp = []
        RF_temp = []
        svm_temp = []
        boosting_temp = []

        for j in range(n): # i = 10

            test_data = df_sub[j]

            S_c = df_sub[0:j] + df_sub[j+1:]
            S_c = list(np.concatenate(S_c, axis = 0))

            train_data = random.sample(S_c, train_sample_size[i])
            common_words = get_most_commons(train_data, skip=100, total=feature_number)
            test_vectors = generate_vectors(test_data, common_words)
            train_vectors = generate_vectors(train_data, common_words)

            y_obs_train = [l[0] for l in train_vectors]

            #get only the xs from the freature matrix
            x_feature_train = [l[1] for l in train_vectors]

            data_matrix_train0 = np.array(x_feature_train)

            data_matrix_train = np.append(data_matrix_train0, np.array(y_obs_train)[:, None], 1)

            y_obs_test = [l[0] for l in test_vectors]

            #get only the xs from the freature matrix
            x_feature_test = [l[1] for l in test_vectors]

            test0 = np.array(x_feature_test)

            test = np.append(test0, np.array(y_obs_test)[:, None], 1)

            #DT
            DT_pred = decision_tree_predict(data_matrix_train, test, 10, 10)

            DT_temp.append(zero_one_loss(DT_pred, test))
            
            #BT
            BT_pred = bagging_pred(data_matrix_train, test, max_depth=10, min_size=10, tree_number=50)

            BT_temp.append(zero_one_loss(BT_pred, test))
            
            #RF
            RF_pred = random_forest(data_matrix_train, test, max_depth=10, min_size=10, tree_number=50)

            RF_temp.append(zero_one_loss(RF_pred, test))
            
            #svm
            svm_new_weights = svm_learn(weights, train_vectors, feature_number, lamda, eta_svm, epsilon, iterator)

            svm_y_pred = svm_classify(svm_new_weights, test_vectors, feature_number)

            svm_temp.append(zero_one_loss2(svm_y_pred, test_vectors))
            
            #Boosting
            boosting_pred = boosting(data_matrix_train, test, max_depth=10, min_size=10, tree_number=50)
            
            boosting_temp.append(zero_one_loss(boosting_pred, test))

            j = j + 1
        
        DT_test_zero_one_loss.append(DT_temp)
        BT_test_zero_one_loss.append(BT_temp)
        RF_test_zero_one_loss.append(RF_temp)
        svm_test_zero_one_loss.append(svm_temp)
        boosting_test_zero_one_loss.append(boosting_temp)

        i = i + 1

    DT_mean = np.mean(DT_test_zero_one_loss, axis=1)
    BT_mean = np.mean(BT_test_zero_one_loss, axis=1)
    RF_mean = np.mean(RF_test_zero_one_loss, axis=1)
    svm_mean = np.mean(svm_test_zero_one_loss, axis=1)
    boosting_mean = np.mean(boosting_test_zero_one_loss, axis=1)


    DT_std = np.std(DT_test_zero_one_loss, axis=1)
    BT_std = np.std(BT_test_zero_one_loss, axis=1)
    RF_std = np.std(RF_test_zero_one_loss, axis=1)
    svm_std = np.std(svm_test_zero_one_loss, axis=1)
    boosting_std = np.std(boosting_test_zero_one_loss, axis=1)



    plt.figure(figsize=(10,5)) 
    plt.subplot(111) 

    plt.title('Means, Std Error of DT, BT, RF, SVM, and boosting Learning Curve by Training Size')
    plt.errorbar(train_sample_size, DT_mean, DT_std, fmt='', color='b', label='DT')
    plt.errorbar(train_sample_size, BT_mean, BT_std, marker='o', linestyle='--', color='r', label='BT')
    plt.errorbar(train_sample_size, RF_mean, RF_std, marker='s', linestyle='--', color='y', label='RF')
    plt.errorbar(train_sample_size, svm_mean, svm_std, marker='x', linestyle='--', color='green', label='SVM')
    plt.errorbar(train_sample_size, boosting_mean, boosting_std, marker='|', linestyle='--', color='black', label='boosting')


    plt.ylabel('zero-one-loss')
    plt.xlabel('TSS')
    plt.legend()
    plt.tight_layout()
    plt.show()


main()
```
